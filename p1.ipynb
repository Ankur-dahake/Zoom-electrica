## What is the cummulative sales volume ( in units) for the first 7 days between 10- 10 -2016 and 16-`10-2016

```python
df_start=df_sale.loc[(df_sale['product_id'] == 7)&(df_sale['Date'] >= '2016-10-10') & (df_sale['Date'] <= '2016-10-16')]
df_start
```

```python
sale_per_day=np.array(df_start['Date'].value_counts())
sale_per_day
```

```python
No_of_days=np.arange(1,8,1)
No_of_days
```

```python
list_of_tuple=list(zip(No_of_days,sale_per_day))
df=pd.DataFrame(list_of_tuple,columns=['No_of_days','sale_per_day'])
df
```

```python
df['cum_sum']=np.cumsum(df['sale_per_day'])
df
```
